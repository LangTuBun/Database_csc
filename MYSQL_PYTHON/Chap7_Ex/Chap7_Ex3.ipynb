```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
# from mylibs import *
import pandas as pd
import time
from mylibs import *

```

Lấy link từng sản phẩm 

```python
url = 'https://babycuatoi.vn/xep-hinh-thong-minh'
driver = webdriver.Edge()
driver.get(url)
link_list = []
# driver.get()
driver.find_element(By.TAG_NAME, value= 'body').send_keys(Keys.END) 
links = driver.find_elements(By.CSS_SELECTOR, value= 'div.product-content a')
for link in links: 
    link_list.append(link.get_attribute('href') + '\n')
print(len(link_list),link_list)
write_txt('babycuatoi.txt', link_list)
driver.close()
```

```python
def get_name():
    name = driver.find_element(By.CSS_SELECTOR, value= 'div.product-info>h1').text
    return name
```

```python
def get_price():
    price = driver.find_element(By.CSS_SELECTOR, value= 'div.box-info.box-price span.price').text
    return price
```

```python
def get_discount():
    discount = driver.find_element(By.CSS_SELECTOR, value= 'div.box-info.box-price label.flash-sale__product-percent').text
    return discount
```

```python
def get_describe():
    list_describe = []
    describe_container = driver.find_element(By.CSS_SELECTOR, value= 'div.detail_cate.css-content')
    describe = describe_container.find_elements(By.TAG_NAME, value= 'p')
    for i in describe:
        list_describe.append(i.text)
        print(i.text)
    return list_describe
```

```python
driver = webdriver.Edge()
product_links = read_txt('babycuatoi.txt')

for i in range(0, len(product_links)):
    driver.get(product_links[i].rstrip('\n'))
    print(i)
    print(get_name())
    print(get_price())
    print(get_discount())
    print(get_describe())
    

driver.close()
```
