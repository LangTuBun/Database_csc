```python
conn = connect('DienMay.db')
for taptin in os.listdir(thu_muc_json):
    duongdan = f'{thu_muc_json}/{taptin}'
    dulieu = read_json(duongdan)
    list_noi_dung = list(dulieu.values())
    list_noi_dung.append(1)
    print(list_noi_dung)
    #thực hiện đưa bảng vào database
    cursor = conn.cursor()
    sql = 'INSERT INTO SanPham(Ten_san_pham, Gia_ban, Gia_goc, Hinh_anh, Ma_danh_muc) VALUES(?,?,?,?,?)'
    cursor.execute(sql, list_noi_dung)
    conn.commit()
    cursor.close()
    print(duongdan)
print('>> Done!')
conn.close()
```
